# Highly unbalanced Dataset

```python
# Because  we have more 0's and very less 1, if we train our model mostly it will predict the 0, it will lean to oneside.
```

```python
# We have only one null value, so it is better to drop it
df.dropna(inplace=True)
```

# Let's balance the dataset

```python
# notedown the mean and standard deviation
# We are going to make the 0 values equal to the 1, but still the mean will be similar
```

```python
df.describe()
```

```python
df.groupby("Class").mean()
```

```python
real = df[df["Class"] == 0]
fake = df[df["Class"] == 1]
```

```python
real["Class"].value_counts(), fake["Class"].value_counts()
```

```python
# Make the real dataset equal to fake

real = real.sample(n=191)
```

```python
fake.shape, real.shape
```

```python
df1 = pd.concat([real,fake],axis=0)  # Axis = 0, will attach the df horizontally
```

```python
df.groupby("Class").mean()
```

```python
df1.groupby("Class").mean()
```

```python
# The mean is similar for both the df
```

```python
X  =df1.drop("Class",axis=1)
y = df1.Class
```

```python
# Split the data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,stratify = y,random_state = 4)   # Stratify will allocate the y values equally
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# Model building
```

```python
from sklearn.linear_model import LogisticRegression
logistc_r = LogisticRegression()
logistc_r.fit(X_train,y_train)
```

```python
# Accuracy for test data

lgc_r = logistc_r.predict(X_test)
accuracy_score(lgc_r,y_test)
```

```python
# Accuracy for train data

lgc_r = logistc_r.predict(X_train)
accuracy_score(lgc_r,y_train)
```

```python
# Both have the similar accuracy score it means the model is performing well and there is no overfitting and underfitting.
# If both are different than it means they are underfit or overfit.
```
